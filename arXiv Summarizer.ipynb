```python
import os
import json
import pandas as pd
from openai import OpenAI

runs = 10
csv_file_path = 'results.csv'
cost_per_input_token = 0.0010
cost_per_output_token = 0.0020

def read_existing_csv(file_path):
    if os.path.exists(file_path):
        return pd.read_csv(file_path)
    else:
        return pd.DataFrame(columns=['Content'])

with open('prompt.txt') as f:
    prompt = f.read()

client = OpenAI()
df_existing = read_existing_csv(csv_file_path)

for run in range(1, runs + 1):
    print(f'Run {run}/{runs}...')
    response = client.chat.completions.create(
        messages=[{
            'role': 'user',
            'content': prompt,
        }],
        model='gpt-3.5-turbo-1106',
        temperature=1,
        )

    content = response.choices[0].message.content
    cost = response.usage.prompt_tokens * cost_per_input_token / 1000 + response.usage.completion_tokens * cost_per_output_token / 1000
    print(f'Cost: ${cost:.2f}\n')
    print(content)
    print('\n')
    
    df_new = pd.DataFrame({'Content': [content]})
    df_existing = pd.concat([df_existing, df_new], ignore_index=True)
    df_existing.to_csv(csv_file_path, index=False)

print('Done!')
```
